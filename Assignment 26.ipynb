Matplotlib is a Python library used for data visualization. It is a comprehensive library that can be used to create a variety of 2D and 3D plots and visualizations. It was created by John D. Hunter in 2003 as a tool to produce publication-quality figures in various formats such as PDF, SVG, PNG, and more.

Matplotlib is used for various purposes such as data exploration, model evaluation, presentation graphics, and more. It is extensively used in data science, machine learning, scientific research, and various other fields where data visualization is essential.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:

Line plot
Scatter plot
Bar plot
Histogram
Pie chart
These are just a few examples of the many different types of plots that can be created using Matplotlib's Pyplot module. Matplotlib provides a wide range of options for customization, which makes it a powerful tool for creating complex visualizations with ease.

```python
#ans 2:
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title("Relationship between X and Y")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#ans 3:
import matplotlib.pyplot as plt
import numpy as np

# data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# create a 2x2 grid of subplots and plot each line in a different subplot
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title("Line 1")

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title("Line 2")

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title("Line 3")

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title("Line 4")

# adjust the spacing between subplots
plt.subplots_adjust(hspace=0.5, wspace=0.5)

# display the plot
plt.show()
```

```python
#ans 4:
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# plot a vertical bar plot
plt.bar(company, profit)
plt.title("Company profits")
plt.xlabel("Company")
plt.ylabel("Profit ($ billions)")
plt.show()

# plot a horizontal bar plot
plt.barh(company, profit)
plt.title("Company profits")
plt.xlabel("Profit ($ billions)")
plt.ylabel("Company")
plt.show()
```

```python
#ans 5:
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

# plot the box plot
plt.boxplot(data)
plt.title("Box plot")
plt.xlabel("Data sets")
plt.ylabel("Values")
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.show()
```
