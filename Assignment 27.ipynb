Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and visually appealing statistical graphics. Here are five popular plots that can be created using Seaborn along with their uses:

Scatter Plot - A scatter plot is used to show the relationship between two continuous variables. It is a useful tool for identifying patterns and trends in data.

Line Plot - A line plot is used to display trends over time or another continuous variable. It is useful for visualizing the changes in a variable over a given time period.

Bar Plot - A bar plot is used to compare different categories or groups. It is useful for visualizing the distribution of a categorical variable.

Heatmap - A heatmap is used to visualize the correlation between two variables. It is useful for identifying patterns and trends in data that may not be immediately visible.

Histogram - A histogram is used to display the distribution of a continuous variable. It is useful for understanding the frequency distribution of a given variable.

```python
#ans 2:
import seaborn as sns
fmri = sns.load_dataset("fmri")
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
```

```python
#ans 3:
import seaborn as sns
titanic = sns.load_dataset("titanic")
sns.boxplot(x="pclass", y="age", data=titanic)
sns.boxplot(x="pclass", y="fare", data=titanic)
```

```python
#ans 4:
import seaborn as sns

# Load the "diamonds" dataset
diamonds = sns.load_dataset("diamonds")

# Create a histogram of diamond prices with different colors for each cut
sns.histplot(data=diamonds, x="price", hue="cut")



```

```python
#ans 5:
import seaborn as sns

# Load the "iris" dataset
iris = sns.load_dataset("iris")

# Create a pair plot with different colors for each species
sns.pairplot(data=iris, hue="species")


```

```python
#ans 6:
import seaborn as sns
flights = sns.load_dataset("flights")
flights_matrix = flights.pivot("month", "year", "passengers")
sns.heatmap(flights_matrix, annot=True, fmt="d")

```
