Let's define the events as follows:
A = an employee is a smoker
B = an employee uses the health insurance plan

We want to find the probability of A given B, i.e., P(A|B). We know that P(B) = 0.7 (given in the problem statement) and that P(A|B) = 0.4 (also given in the problem statement). We need to find P(B|A), i.e., the probability that an employee uses the health insurance plan given that they are a smoker.

We can use the formula as follows:

P(B|A) = P(A|B) * P(B) / P(A)

To find P(A), we need to use the law of total probability:

P(A) = P(A|B) * P(B) + P(A|B') * P(B')

where B' is the complement of B (i.e., an employee does not use the health insurance plan). We know that P(A|B') = 0 (since non-users of the health insurance plan cannot be smokers), so we can simplify the equation as:

P(A) = P(A|B) * P(B)

Now we can substitute the values we have:

P(B|A) = P(A|B) * P(B) / P(A)
= 0.4 * 0.7 / (0.4 * 0.7)
= 0.4

Therefore, the probability that an employee is a smoker given that he/she uses the health insurance plan is 0.4 or 40%.

Feature Independence: Both algorithms assume that the features (words) in a document are independent of each other. However, Bernoulli Naive Bayes assumes that the absence or presence of a feature is independent of the presence or absence of other features, while Multinomial Naive Bayes assumes that the occurrence of a feature is independent of the occurrence of other features, given the class variable.

Application: Bernoulli Naive Bayes is suitable for text classification tasks where the goal is to distinguish between two classes, such as spam vs. non-spam emails. Multinomial Naive Bayes is commonly used for text classification tasks where the goal is to distinguish between more than two classes, such as sentiment analysis or topic classification.

In the one-vs-all approach, we build one binary classifier for each class, where the positive class is the class of interest and the negative class includes all other classes. During classification, each classifier predicts the probability that the input belongs to its positive class, and the class with the highest probability is chosen as the final prediction.

For example, suppose we have a dataset with three classes A, B, and C. To classify a new instance, we would train three Gaussian Naive Bayes classifiers: one to distinguish A from B and C, one to distinguish B from A and C, and one to distinguish C from A and B. Each classifier would output a probability that the instance belongs to its positive class, and we would choose the class with the highest probability as the final prediction.

It's important to note that the one-vs-all approach can be less accurate than other multi-class classification algorithms such as multinomial Naive Bayes or logistic regression, especially when the classes are imbalanced or there is overlap between the classes. However, Gaussian Naive Bayes with the one-vs-all approach can still be a useful and simple method for multi-class classification, especially when the number of classes is small.

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('spambase.data')
df.head()
```

```python
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB
from sklearn.model_selection import cross_val_score


target,data = df


# Bernoulli Naive Bayes
bnb = BernoulliNB()
bnb_scores = cross_val_score(bnb, data, target, cv=10, scoring=['accuracy', 'precision', 'recall', 'f1'])

# Multinomial Naive Bayes
mnb = MultinomialNB()
mnb_scores = cross_val_score(mnb, data, target, cv=10, scoring=['accuracy', 'precision', 'recall', 'f1'])

# Gaussian Naive Bayes
gnb = GaussianNB()
gnb_scores = cross_val_score(gnb, data.toarray(), target, cv=10, scoring=['accuracy', 'precision', 'recall', 'f1'])

# Report the performance metrics for each classifier
print("Bernoulli Naive Bayes:")
print("Accuracy:", bnb_scores['test_accuracy'].mean())
print("Precision:", bnb_scores['test_precision'].mean())
print("Recall:", bnb_scores['test_recall'].mean())
print("F1 Score:", bnb_scores['test_f1'].mean())

print("\nMultinomial Naive Bayes:")
print("Accuracy:", mnb_scores['test_accuracy'].mean())
print("Precision:", mnb_scores['test_precision'].mean())
print("Recall:", mnb_scores['test_recall'].mean())
print("F1 Score:", mnb_scores['test_f1'].mean())

print("\nGaussian Naive Bayes:")
print("Accuracy:", gnb_scores['test_accuracy'].mean())
print("Precision:", gnb_scores['test_precision'].mean())
print("Recall:", gnb_scores['test_recall'].mean())
print("F1 Score:", gnb_scores['test_f1'].mean())
```
